```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
df.info()
```

```python
missing_values = df.isnull().sum()
print("\nMissing Values per Column:\n", missing_values)
```

```python
median_ages = df.groupby('Sex')['Age'].median()

df['Age'] = df.apply(
    lambda row: median_ages[row['Sex']] if pd.isnull(row['Age']) else row['Age'],
    axis=1
)
```

```python
df['Title'] = df['Name'].str.extract(',\s*([^.]*)\.', expand=False)
```

```python
df['Title'] = df['Title'].replace({'Mlle': 'Miss', 'Ms': 'Miss', 'Mme': 'Mrs'})
```

```python
rare_list = ['Dr', 'Rev', 'Col', 'Major', 'Countess', 'Lady', 'Jonkheer',
             'Don', 'Dona', 'Capt', 'Sir']

df['Title'] = df['Title'].apply(lambda t: 'Rare' if t in rare_list else t)
```

```python

title_median_age = df.groupby('Title')['Age'].median()

df['Age'] = df.apply(
    lambda row: title_median_age[row['Title']] if pd.isnull(row['Age']) else row['Age'],
    axis=1
)
```

```python
print("\nSurvival Rate by Title:\n")
print(df.groupby('Title')['Survived'].mean())

# Final missing values count
print("\nFinal Missing Value Check:\n", df.isnull().sum())
```
